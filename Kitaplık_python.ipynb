#Kütüpaneler

---
Kütüphaneler, daha önceden yapılmış işlemleri saklyan ve bunları tekrar kullanabilmemizi sağlayan bir yapıdır. Kütüphanler ücretsiz bir şekilde kullanılabilir olabileceği gibi paralı da olabilir. Bunları yazan kişiler, tekrar tekrar çağırabilmemiz için erişime açık isimler kullanarak oluşturmuşlardır. Milyarlarca kütüphane vardır. Örneğin: <br>

* Rastgele sayı ve içerik üretmemizi sağlayan **random** kütüphanesi
* işletim sisteminde yer alan klasörlerle işlem yapmamızı sağlayan
**os** kütüphanesi
* Matemetiksel üst düzey işlemler yapmamızı sağlayan math kütüphanesi
* Oyun geliştirmek için **turtle** ve **pygame** kütüphaneleri
* Arayüzü program tasarlamamızı sağlayan **tkinter** kütüphanesi
* Veritabanı araçlarını kullanabilmeyi sağlayan **sqlite** kütüphanesi

##Peki Kütüphaneler Nasıl Kodlarımızda Kullanılır?
Kütüphaneleri kullanmak için öncelikle kod metnimizin en tepesinde onları çağırmamız gerekir. Eğer bir kütüphane bilgisayarımızdayüklü ise kütüphane hatasız çalışır. Kütüphane kodumuzu eklemek için **import** deyimini kullanıyoruz. Örneğin: <br>



```
import random
import os
import math
```




````python
#Kütüphanlerli programın başında çağırıyoruz.
import random #Rastgele içerik üretmeyi sağlar.

rastgele_sayi = 0

#Rastgele sayıyı random kütüphanesinden çağırmak istiyorsam, kütüphanenin adını yazrım.
#Ardından random nokta işareti ile istediğim işlemi çağırırım.
rastgele_sayi = random.randint(1,5)
#randint -> girilen iki değer arasında rastgele sayı üretmemizi sağlar.
print(rastgele_sayi)
if rastgele_sayi == 1:
  print("Ajan 49")
if rastgele_sayi == 2:
  print("ZAY")
if rastgele_sayi == 3:
  print("sigma müfit")
if rastgele_sayi == 4:
  print("Beraga")
if rastgele_sayi == 5:
  print("Üstat Mareşel 5.Lider Emir Yiğit Durmuş")
````

**NOT:** Kütühaneleri çapırırken kısaltma isimlerle de çağırabiliriz.
Bunun için **as** takısı kullanılır.
**Örenğin;** import random as r

````python
import random as rd #Kısaltma yapılabiliyor.
import math as m

sayi1 = rd.randint(1, 10) #1-10 arasında rastgele sayıyı değişkeni atabiliyoruz.

print(sayi1)

#Matematiksel Fonksiyonlar
print(m.factorial(sayi1)) #Faktöriyel
print(m.comb(sayi1, 2)) #Kombinasyon (iki dğer alır) (sayi1, sayi2)
print(m.perm(sayi1, 2)) #Permitasyon
print(m.sqrt(sayi1)) #Karekök alma
print(m.floor(sayi1)) #Yuvarlama
print(m.pow(sayi1, 4)) #İstediğin sayının Kökkuvvetini alma
print(m.log(sayi1, 3)) #İstediğin tabandan
print(m.pi) #Pi sayısı
print(m.e) # e sayısı
````
