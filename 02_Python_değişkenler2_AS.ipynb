#Kullanıcıdan Değer Alma
**Input Komutu:** Bu input komutu, kullanıcıdan veri almamızı sağlar.
Aldığı verileri metin değerinde saklar!<br>

Sayıları almak için değişkeni **int()** parantezine almamız gerekir!

```python
#Sayı değişkenlerini kullıcıdan alma:
yas = input("Yaşınızı Giriniz:")

print("Yaşınız:"+" "+str(yas)) # + işareti metinleri birleştirir. Dolayısıyla yanına sayı yazamazsınız.
print("Yaşınız:",int(yas)) # , işareti proglamlama nesnelerini yan yana yazabilmeyi sağlar. Sayı ve metin bir arada yazılabilir.
print("Yaşınız:"+" "+str(int(yas))) # İç içe str int yapılabilir.
print(int(yas))
```

#Değişkenleri İşlemler
String yani metin değişkenlerinin özgü işlemleri, sayıların kendine özgü ayrı işlemleri vardır. İlk olarak sayı değişkenleri ile işlemlere başlayalım.

**NOT:** "print'te" + işareti aynı türden değerleri yan yana yazar veya toplar.
,(virgül) işareti farklı  türden nesneleri ayrı olarak algılar ve yan yana yazabilmemizi sağlar.





```python
#Sayı değişkenin işleme sokma işlemi:
brinci_sayi = input("İlk sayıyı giriniz:")
ikinci_sayi = input("İkinci sayıyı giriniz:")
#NOT: Genelde bir işlem yapılıyorsa o işlemi ayri bir değişkene yazarız.
print("Toplamı:",int(brinci_sayi)+int(ikinci_sayi))
print("Çıkarma:",int(brinci_sayi)-int(ikinci_sayi))
print("Çarpma:",int(brinci_sayi)*int(ikinci_sayi))
print("Bölme:",int(brinci_sayi)/int(ikinci_sayi))
# Değişkenleri integer'a dönüştür
brinci_sayi = int(brinci_sayi)
ikinci_sayi = int(ikinci_sayi)
toplam = brinci_sayi + ikinci_sayi
cikarma = brinci_sayi - ikinci_sayi
carpma = brinci_sayi * ikinci_sayi
bolme = brinci_sayi / ikinci_sayi
print(str(brinci_sayi) + "ile" + str(ikinci_sayi) + " arasında 4 işlem şu şekildedir:")
print("Toplamı:", toplam)
print("Çıkarma:", cikarma)
print("Çarpma:", carpma)
print("Bolme:", bolme)
```

#Değişkenlerle Sayısal İşlemler
Metin işlemlerinden farksızdır. Sadece önemli olan değişkenlerin int mi yoksa str mi olduğunu kontrol ederek işlem yapmamız gerekir.
int -> Integer -> Sayılar
str -> String -> Metin

```python
#Basit dört işem yapmak
#input her zaman string yani metin türünde değerler alır. Sayıya çevirmek için int parantezini alırım.
sayi1 = int(input("Birinci sayıyı giriniz:")) #Metini sayıya cevirmek için input'u int parantezine alırım
sayi2 = int(input("İkinci sayıyı giriniz:"))
#toplam = 0
#cikarma = 0
toplam,cikarma,carpma,bolme = 0, 0, 0, 0 #Değişkenleri yan yana değerleri ile birlikte yazabilirim
toplam = sayi1 + sayi2
cikarma = sayi1 - sayi2
carpma = sayi1 * sayi2
bolme = sayi1 / sayi2
print("Toplamı:",toplam)
print("Çıkarma:",cikarma)
print("Çarpma:",carpma)
print("Bölme:",bolme)

```
