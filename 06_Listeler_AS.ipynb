#Listeler
Değişkenler bir tane veri saklayan  ve tek amaçla olışturulan yapılardı.
Listeler ise birden fazla veriyi içinde saklayan ve farklı türlerden
verileri kendi içinde tutabilen yapılardır. <br>

Liste oluşturmak oldukça basittir. Öncelikle listenin adı verilir ve eşittir işareti konulur ve ardından köşeli parantezler [] arasına değerler yazılır.

```python
sehirler = ["Moskova", "Paris", "St.Petersburg", "Ankara"]
haftanin_gunleri = ["Pazartesi", "Sali", "Çarşamba", "Perşembe", "Cuma", "Cumartesi", "Pazar"]
rastgele_liste = ["Elma", 12, "Bilinmez", "Çorba"]
print(sehirler)
print(haftanin_gunleri)
print(rastgele_liste)
```

```python
maker = ["Müfit Führer", "Zeynep KPop", "Ataberk Negro", "Yiğit Negro", "Uğur Kratos"]
print(maker[1])
```

**Not:** Listelerde bir elemanı içinden çağırmak istiyorsak listenin adını yazıp yanına köşeli parantez [ ] açıp içerisine rakam yazıyoruz. Bu rakam 0'dan başlar ve listenin eleman sayısnın 1 eksiğine kadar gider. Mesela haftanın günleri 7'yse listedeki elemanları 0 ve 6 arasında yazdırabilirim.

Böylece ilk eleman 0'dır. Bu köşeli parantez içindeki değerlere **listenin indisi** denir.

```python
import random
listem = ["elma", "armut", "ayva",123,True]
rastgele_indis = random.randint(0,len(listem)-1) #indis sayısı eleman sayısından 1 eksil olmalı
print(listem[rastgele_indis])
```

#Listelere Input ile Eleman Ekleme ve Çıkarma
Input kullanarak listelere eleamn ekleyip çıkarabilriz. Bazı programlama dillerinde add (ekleme) vaye append (listeye içine ekleme) anlamına gelen bir fonksiyon ile ekleme yapılabilir. <br>

Çıkarma işemlerinde delete (silme) veya remove (listeden çıkarma) anlamında bu fonksiyonlar kullanılabilir. İkisinden örnek vermek gerekirse liste.append (sayi) veya liste.remove(3) gibi farklı sekillerde kullanılabilir.

```python
tatlilar = ["triliçe","kemal paşa","sütlaç","puding","baklava","pasta","kadayıf"]
tatli = input("Bir tatlı ismi giriniz: ")
tatlilar.append(tatli)
print(tatlilar)
#Listeden ekleneni silebilirim.
tatlilar.remove(tatli)
print("Silme işlemi")
print(tatlilar)
silinecek = input("Silmek istediğiniz tatlıyı giriniz: ")
tatlilar.remove(silinecek)
print("ekstra silmeden sonra:")
print(tatlilar)
```

#Listelerde Farklı İşlemler
Listeleri temizleme, eleman seçip aynı anda çıkarma, listeyi tersten yazdırma, liste eleman sayısı veya indis bilgisi alma gibi işlemler yapabliriz.

```python
# clear komutu -> her şeyi sild
haftanin_gunleri = ["Pazartesi", "Sali", "Çarşamba", "Perşembe", "Cuma", "Cumartesi", "Pazar"]
haftanin_gunleri.clear()
print(haftanin_gunleri)
```

```python
#Count komutu -> Yazılan Elemanı sayar
haftanin_gunleri = ["Pazartesi", "Sali", "Çarşamba", "Perşembe", "Cuma", "Cumartesi", "Pazar"]
gun = haftanin_gunleri.count("Sali")
print(gun)
```

```python
#İndex komutu -> yazılan elemanın kaçıncı index olduğunu yazar.
haftanin_gunleri = ["Pazartesi", "Sali", "Çarşamba", "Perşembe", "Cuma", "Cumartesi", "Pazar"]
print(haftanin_gunleri.index("Çarşamba"))
```

```python
#pop komutu -> listenin içinden bahsedilen index'li elemanları alır yazar ve listeden çıkarır
haftanin_gunleri = ["Pazartesi", "Sali", "Çarşamba", "Perşembe", "Cuma", "Cumartesi", "Pazar"]
gun = haftanin_gunleri.pop(3)
print(gun)
print(haftanin_gunleri)
```

```python
#Reverse komutu-> Listeyi ters çevirir.
haftanin_gunleri = ["Pazartesi", "Sali", "Çarşamba", "Perşembe", "Cuma", "Cumartesi", "Pazar"]
haftanin_gunleri.reverse()
print(haftanin_gunleri)
```

#Listelerde rama yapma ve döngü kullanımı
Listeler içerisinde belirli koşullar sağlandığında yazdırılacak veya işe yarayacak bilgileri alabilmek için listelerde arama yapılmalıdır.
Listelerde arama yapmak için döngüler kullanılbilir.

```python
#0 - 250 arasında 3'er 3'er sayıları yazdırma
ucerli_liste = []
onbesin_kati = []
for a in range(0,250,3):
  ucerli_liste.append(a)
print(ucerli_liste)
print("Toplam eleman sayısı",len(ucerli_liste))
# Bu listenin içinden 5'e bölünen sayıları ekrana yazdırma
for eleman in range(0, len(ucerli_liste)-1):
  if ucerli_liste[eleman] % 5 == 0:
    onbesin_kati.append(ucerli_liste[eleman])
print(onbesin_kati)
print(ucerli_liste[eleman])
```
