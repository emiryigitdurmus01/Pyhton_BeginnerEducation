#Değişkenler
Yazılımda içinde veri sakladığımız yapılara değişken diyoruz. Bu veri yazılımcının veya kullanıcının saklayabileceği önemli bilgilerdir.


```python
#Bir değişken oluşturalım.
#Değişkeni benim adımı soyadımı saklasın.
isim = "Emir"
soyad = "Durmuş"
AdSoyad = "Emir Yiğit Durmuş"
print(isim)
print(soyad)
print(AdSoyad)
print(isim+" "+soyad)
```

#Değişkeni Oluştururken Dikkat!


1.   Değişken ismini değikenin amacına göre içinde barındırdığı bilgiye göre vermemiz gerekir.
2.   Değişken isimlerini oluştururken Türkçe karakter kullanılmaz
3.   Değiken oluşturuken isimler arasına booşluk bırakmadan yazarız
  Ad Soyad (**Yanlış**)
  Ad_Soyad (**Doğru**)






```python
#Değişken Türleri
#Bir Arabanın Özellikleri
araba_marka="Toyota Corola" #Metin -> string (str) demektir
araba_hiz= 4000 #Sayi -> Integer (int) demektir
arabanin_yuksekligi= 139.9 #Yukseklik -> Float (f) demektir
arabanın_genisligi = 250,3 #Genişliği -> Double (d) demektir
#Değişken Türleri
#Bir Arabanın Özellikleri
araba_marka="Toyota Corola" #Metin -> string (str) demektir
araba_hiz= 4000 #Sayi -> Integer (int) demektir
arabanin_yuksekligi= 139.9 #Yukseklik -> Float (f) demektir
arabanın_genisligi = 250.3 #Genişliği -> Double (d) demektir
arabanin_calisiyor_mu = True #Doğru/Yanlış -> Boolean (bool) demektir
araba_pahali_mi = False #Doğru = True, Yanlış = False
print(araba_marka+"\n"+str(araba_hiz)+"\n"+str(arabanin_yuksekligi)+"\n"+str(arabanın_genisligi)+"\n"+str(arabanin_calisiyor_mu)+"\n"+str(araba_pahali_mi)+"\n")
```

#Değişkenlerle İşlemler
Değişkenleri içindeki değerden bağımsız olarak işleme sokabiliriz.
Sayıları toplayabiliri metinleri yan yana veya alt alta yadırabiliriz.
İçindeki değer amacı doğrultusunda kullanılmalıdır! Değişkenleri işleme sokmak için yeni bir değişken de oluşturulablir.

```python
sayi1 = 23
sayi2 = 5e+2
toplam = sayi1+sayi2
print(toplam)
print(str(sayi1)+"+"+str(sayi2)+"="+str(toplam))
```
