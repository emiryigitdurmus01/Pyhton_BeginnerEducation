**len()** -> lenth in türkçesi uzunluktur. Bir metnin kaç karakter olduğunu bize söyler.

```python
print('Atatürk\'ün dediği gibi "Egemenlik kayıtsız şartsız milletindir."')
print('''Atatürk'ün dediği gibi "Egemenlik kayıtsız şartsız milletindir."''')
```

**NOT:** Başina ve sonuna 3 tane tek tırnak koyarsak içinde özel karakterleri çalıştırabiliriz. (?,"";'')

**Kaçış Dizisi(\\):** Kaçış dizisi özel kartakter tanımlamak için kullanılan bir kavramdır.

```python
print((8-12),(8+12),(8*12),(8/12))
print((8-12)+(8+12)+(8*12)+(8/12))
```

```python
print("Bugün","Hava","Çok","Güzel")
print("Bugün"+"Hava"+"Çok"+"Güzel")
```

**NOT:** Print kmotunda sayılarda virgül ile yan yana işlemler sırasıyla yaptırabilir. Ancak metinde virgül ve artı işareti şle yan yana yazdırabilir.

```python
print("Bugün","Hava\n","Çok","Güzel")
```

**\n** Alt satıra yazıyı indir.
**\r**Kendisinden öneckileri siler.
**\t**Kendisinden sonraya 4 boşluk bırakır.

```python
print("Bugün\r","Hava","Çok","Güzel")
print("Bugün","Hava","Çok\n","Güzel")
```

```python
print("Bugün\t","Hava","Çok","Güzel")
print("Bugün","Hava","Çok\t","Güzel")
```
